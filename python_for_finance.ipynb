```python
##this program predicts stock prices by using machine learning models

import quandl
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
##get the stocks data--------->
df = quandl.get("WIKI/AMZN")
df.head()
```

```python
##get the adjusted close price----->
df = df[['Adj. Close']]
df.head()
```

```python
## a variable for predicting 'n' days out into the future
forecast_out = 30
##create another column shifted 'n' units up
df['Prediction'] = df[['Adj. Close']].shift(-forecast_out)

##print the new data set
df.tail()
```

```python
##created the independent data set (x)
#convert the dataframe to a numpy array
X = np.array(df.drop(['Prediction'], axis=1))  # Ensure 'axis' is specified

##remove the last n rows
X = X[:-forecast_out]
print(X)
```

```python
##create the dependent dataset (y)
#convert the dataframe to a numpy array (All of the values including the NAN's)
y= np.array(df['Prediction'])
## get all the y values except the last n rows
y=y[:-forecast_out]
print(y)
```

```python
##split the dataset
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#create and train the svm
svr_rbf = SVR(kernel='rbf',C=1e3,gamma=0.1)
svr_rbf.fit(X_train,y_train)
```

```python
##testing model:score returns the cofficient of determination R^2 of the prediction
##the best possible score is 1.0
svm_confidence = svr_rbf.score(X_test,y_test)
print('svm confidence:',svm_confidence)
```

```python
##create and train the linear_regression model
lr= LinearRegression()
lr.fit(X_train,y_train)
```

```python
##testing linear model->
lr_confidence = lr.score(X_test,y_test)
print('lr confidence:',lr_confidence)
```

```python
## set x_forecast equal to  last 30 rows of the original data set from Adj. Close column
x_forecast = np.array(df.drop(['Prediction'],axis=1))[-forecast_out:]
print(x_forecast)
```

```python
## print linearRegression and svm model predictions for the next n days
lr_prediction = lr.predict(x_forecast)
print(lr_prediction)

svm_prediction = svr_rbf.predict(x_forecast)
print(svm_prediction)
```
